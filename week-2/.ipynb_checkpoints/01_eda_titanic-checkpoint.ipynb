```python
import pandas as pd

#Loading the Data
df = pd.read_csv("titanic.csv")

#Cleaning The Data
df = df.fillna(
    {"Age": df["Age"].median()}
    ) #means replace the NaN vaules with 28(middle value): {"Age" : 28}
df = df.drop(columns=["Cabin"])
df = df.dropna(subset=["Embarked"])

#Calculations(compute)
Survived_ByGender = df.groupby("Sex")
rate = Survived_ByGender["Survived"].mean() #if you use count() it will give you the whole people where in the ship but if you use sum() it'll give you the number of survivors
rate.to_csv("Cleaned _Titanic")
```
